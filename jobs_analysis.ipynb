*****Load and explore:*****

```python
df = pd.read_csv('../data/jobs_in_data.csv')
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

*****Check basic stats:*****

```python
df.info()
print("\nMissing values:", df.isnull().sum().sum())
```

*****Here's what I have:*****

*****Average salary by experience level:*****

```python
exp_salary = df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
print(exp_salary)

exp_salary.plot(kind='bar', figsize=(8,5), color='steelblue')
plt.title('Average Salary by Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

*****Top 10 highest paying job titles:*****

```python
top_jobs = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
print(top_jobs)

top_jobs.plot(kind='barh', figsize=(10,6), color='coral')
plt.title('Top 10 Highest Paying Data Jobs')
plt.xlabel('Average Salary (USD)')
plt.tight_layout()
plt.show()
```

*****Remote vs onsite salary:*****

```python
work_salary = df.groupby('work_setting')['salary_in_usd'].mean().sort_values(ascending=False)
print(work_salary)

work_salary.plot(kind='bar', figsize=(8,5), color='green')
plt.title('Average Salary by Work Setting')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

*****Key insights from the data:*****

*****Jobs by category:*****

```python
job_counts = df['job_category'].value_counts().head(10)

job_counts.plot(kind='barh', figsize=(10,6), color='purple')
plt.title('Most Common Data Job Categories')
plt.xlabel('Number of Jobs')
plt.tight_layout()
plt.show()
```
